```python
def init_driver():
    driver = webdriver.Chrome()
    driver.get(url_thiriet)
    return driver

def select_country():
    
    country = driver.find_element_by_id('pays_id')
    select_country = Select(country)
    select_country.select_by_value('F')
    
def post_code(p):
    
    try:
        post = driver.find_element_by_id("cp")
        
    except NoSuchElementException:
        post_code(p)

        
    post.clear()
    post.send_keys(str(p))
    post.send_keys(Keys.RETURN)
        
    return post
        
def select_location(loc):
    
    try:
        town = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID,'ville_code_insee_iris')))
        town = driver.find_element_by_id("ville_code_insee_iris")
        
    except NoSuchElementException:
        select_location(loc)
        
    town_drp = Select(town)
    town_drp.select_by_visible_text(loc)
    
    print(loc)
    
def press_select():
    
    submit_button = driver.find_element_by_xpath('/html/body/section/div/div[3]/div[2]/div[3]/form/div[2]/div/button')
    submit_button.click()
    
    
    
    
```

```python
import threading
import time

from openpyxl import load_workbook
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
from openpyxl.styles import Color, PatternFill, Font, Border
from openpyxl.styles import colors
from openpyxl.cell import Cell

from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.select import Select

url_thiriet = 'https://livraison.thiriet.com/accueil/mes-services/mes-services,171,1149.html?&args=Y29tcF9pZD0xNjUmYWN0aW9uPWxpc3RlJmlkPSZjaGFuZ2VyX3ZpbGxlPTF8'

list_thiriet = 'brief/list-thiriet2.xlsx'
df_thiriet = pd.read_excel(list_thiriet, dtype=str)

postal_thiriet = df_thiriet['postal_code_thiriet']

city_thiriet = df_thiriet['commune_nom']
city_thiriet_new = []

for cit in city_thiriet:
    #cit = str(cit)
    cit = cit.upper()
    cit = cit.replace('-',' ')
    cit = cit.replace('SAINT','ST')
    print(cit)
    city_thiriet_new.append(cit)
    
```

```python
def select_city(a):
    
    #town = browser.find_element_by_id("ville_code_insee_iris")
    
    try:
        #town = EC.presence_of_element_located((By.ID, "ville_code_insee_iris"))
        #WebDriverWait(browser, 25).until(town)
        #town = WebDriverWait(browser, 25).until(EC.presence_of_element_located(browser.find_element_by_id("ville_code_insee_iris")))
        
        town = WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.ID, 'ville_code_insee_iris')))
        town_drp = Select(town)
        name_town = a
        town_drp.select_by_visible_text(name_town)
        print(a)
        #time.sleep(3)
        
    except TimeoutException:
        print("Town not found")
        
        
def find_result():
    
    try:
        link = WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.XPATH, "/html/body/section/div/div[4]/div/form/div/div[1]/div")))
        return link.text
    
    except TimeoutException:
        
        try:
            link = WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.XPATH,'//*[@id="cc-container"]/div/div[1]/div/p')))
            return link.text
    
        except TimeoutException:
            
            try:
                link = browser.find_element_by_xpath('/html/body/section/div/div[4]/div/div/div/div')
                return link.text
            except TimeoutException:
                #link = "Value Error"
                pass
                
            
            
        
    #return link.text
            
        
    
    

def browsing_activity(val, num, city):
    
    time.sleep(2)
    
    post = browser.find_element_by_id("cp")
    
    post.clear()
    #post.send_keys(str(num))
    post.send_keys(num)
    post.send_keys(Keys.RETURN)

    print(str(num))
   
    
    #town_button.click()
    
    time.sleep(2)
    
    thread = threading.Thread(target=select_city(city))
    
    thread.start()
    
    thread.join()
    
    
    submit_button = browser.find_element_by_xpath('/html/body/section/div/div[3]/div[2]/div[3]/form/div[2]/div/button')
    #submit_button = WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.XPATH,'/html/body/section/div/div[3]/div[2]/div[3]/form/div[2]/div/button')))
    submit_button.click()
    
    
    
    #message = browser.find_elements_by_xpath('/html/body/section/div/div[2]/div/div/p')[0].text
    #link = WebDriverWait(browser, 15).until(EC.presence_of_element_located((By.XPATH, "/html/body/section/div/div[4]/div/form/div/div[1]/div")))
    #link = browser.find_elements_by_xpath("/html/body/section/div/div[4]/div/form/div/div[1]/div")
    message = find_result()
   
    browser.back()
    browser.back()
    
    time.sleep(2)
    
    post2 = browser.find_element_by_id("cp")
    post2.clear()
    
    time.sleep(1)
    
    title = browser.find_element_by_xpath('//*[@title="Je sélectionne ma ville"]')
    title.click()
    #time.sleep(10)
    
    
    
    #browser.close()
    
    return message

def updating_excel(index, pos, area, recep):
    wb = load_workbook('brief/list-thiriet2.xlsx')
    
    ws = wb['Communes-Export-2018-September-']
    
    redFill = PatternFill(start_color='FFFF0000',end_color='FFFF0000',fill_type='solid')
    
    cell_1 = 'F%d'  % (index + 1)
    ws[cell_1] = pos
    #ws[cell_1].fill = redFill
    #print(index)
    
    cell_2 = 'G%d'  % (index + 1)
    ws[cell_2] = area
    #ws[cell_2].fill = redFill
    
    cell_3 = 'H%d'  % (index + 1)
    ws[cell_3] = recep
    ws[cell_3].fill = redFill

    wb.save('brief/list-thiriet2.xlsx')

    
    
    
    
```

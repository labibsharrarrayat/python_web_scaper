```python
from bs4 import BeautifulSoup
import requests
URL = 'https://livraison.thiriet.com/accueil/mes-services/mes-services,171,1149.html?&args=Y29tcF9pZD0xNjUmYWN0aW9uPWxpc3RlJmlkPSZ8'
```

```python
content = requests.get(URL)
soup = BeautifulSoup(content.text,'html.parser')
```

```python
soup
```

```python
rows = soup.find_all('div')
for r in rows:          # Print all occurrences
    print(r.get_text())
```

```python

divTag = soup.find_all("div", {"class": "title"})

for tag in divTag:
    tdTags = tag.find_all("i", {"class": "pictos-livraison_domicide"})
    print(tag)
    for tag in tdTags:
        print(tag.text)
```
